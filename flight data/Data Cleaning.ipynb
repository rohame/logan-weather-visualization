```python
import pandas as pd
data_files = ['201602','201603','201604','201605','201606','201607','201608','201609','201610','201611','201612',
             '201701','201702','201703','201704','201705','201706','201707','201708','201709','201710','201711','201712',
             '201801','201802','201803','201804','201805','201806','201807','201808','201809','201810','201811','201812']
data = pd.read_csv('201601.csv')
for file in data_files:
    data.append(pd.read_csv(file+'.csv'),ignore_index=True)
data = data.drop(columns=['Unnamed: 14'],axis=1)
```

```python
data = data[(data['ORIGIN']=='BOS')]
```

```python
data[data['DEP_DELAY']<0] = 0
```

```python
data.head(5)
```

```python
data_mean = data.groupby('FL_DATE').mean()
data_out = data_mean[['DEP_DELAY','CANCELLED','WEATHER_DELAY']].drop([0])
data_out['DEP_DELAY'] = round(data_out['DEP_DELAY'],2)
data_out['CANCELLED'] = round(data_out['CANCELLED'],2)
data_out['WEATHER_DELAY'] = round(data_out['WEATHER_DELAY'],2)
```

```python
data_out.head(5)
```

```python
data_out.to_csv('data.csv')
```
